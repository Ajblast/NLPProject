# MORE TWEETS

```python
from tweetscrape import *
from dotenv import load_dotenv
from os import getenv
import pandas as pd

load_dotenv() # Load the env variables stored in local .env file
twitter_oauth_ian  = getenv("TWITTER_OAUTH_IAN")
```

```python
archive_scraper = ScraperArchive(twitter_oauth_ian)
```

## Puppies

```python
query = '(puppy OR puppies)'

puppies_df = archive_scraper.scrape(query=query, max_results=15_000, lang='en')
puppies_df.to_csv("puppies_15k.csv")

pagination_token = archive_scraper.get_pagination_token()
print(f"Pagination token: {pagination_token}")
```

## Supreme Court Tweets

```python
archive_scraper.params['start_time'] = '2022-01-01T00:00:00Z'
archive_scraper.params['end_time'] = '2022-11-28T00:00:00Z'
archive_scraper.params
```

```python
query = '(supreme OR court) (#supremecourt OR #scotus)'
```

```python
sc_tweets = archive_scraper.scrape(query, max_results=10_000, lang='en')
```

```python
query_no_hashtags = '(supreme court) (US OR USA OR United States)'
more_sc_tweets = archive_scraper.scrape(query, max_results=10_000, lang='en')
```

```python
sc_tweets_all = pd.concat([sc_tweets, more_sc_tweets], ignore_index=True)
sc_tweets_all.shape
```

```python
sc_tweets_no_dup = sc_tweets_all.drop_duplicates(subset=['text'])
sc_tweets_no_dup.shape
```

```python
archive_scraper.get_pagination_token()
```

```python
query_no_retweet = '(supreme court) (US OR USA OR United States) -is:retweet'
no_retweets_10k = archive_scraper.scrape(query_no_retweet, max_results=10_000, lang='en', pagination_token='b26v89c19zqg8o3fpzhlr3e56l7pl70wy9hx2bm93g1a5')
```

```python
merged_sc_tweets = pd.concat([sc_tweets_no_dup, no_retweets_10k], ignore_index=True)
```

```python
merged_sc_tweets.drop_duplicates(subset=['text'], inplace=True)
merged_sc_tweets.shape
```

```python
merged_sc_tweets.to_csv("2022-supreme-court-tweets-15k.csv")
```
